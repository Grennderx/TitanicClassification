```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df
```

```python
df_test
```

```python
df_train = pd.concat([df, df_test], ignore_index=True)
df_train
```

```python
df_columns = df_train.columns
for i in df_columns:
  print(f'{i}: {df_train[i].isna().sum()} пустых значений')
```

```python
# слишком много отсутствующих значений признака Cabin - удалим признак
df_train = df_train.drop('Cabin', axis=1)
```

```python
df_train
```

```python
#удалим две строчки с пустым признаком Embarked
df_train.dropna(subset='Embarked', inplace=True)
df_train.dropna(subset='Fare', inplace=True)
df_train
```

```python
#возьмем столбец с ID для финального вывода и удалим его
pass_id = df_train['PassengerId'][df_train['Survived'].isna()]
df_train = df_train.drop('PassengerId', axis=1)

#удалим столбец с названием билета Ticket и именем Name
df_train = df_train.drop('Ticket', axis=1)
df_train = df_train.drop('Name', axis=1)
pass_id
```

```python
df_train['Sex'] = df_train['Sex'].map({'male': 0, 'female': 1})
df_train
```

```python
df_train['Embarked'].value_counts()
```

```python
df_train = pd.get_dummies(df_train)
```

```python
#предскажем пропуски признака age на основе имеющихся данных
df_train_age = df_train.copy()
df_train_age.dropna(subset='Age', inplace=True)
df_pred_age = df_train[df_train['Age'].isna()]

df_train_age = df_train_age.drop('Survived', axis=1)
df_pred_age = df_pred_age.drop('Survived', axis=1)

#распределим данные для age-целевой переменной
X_train = df_train_age.drop('Age', axis=1)
X_test = df_pred_age.drop('Age', axis=1)
y_train = df_train_age['Age']
y_test = df_pred_age['Age']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
scaler.fit(X_test)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = np.round(abs(model.predict(X_test)))
print(y_pred)
```

```python
#заменяем пропущенные значения возраста предсказанными
df_train.loc[df_train.Age.isna(), 'Age'] = y_pred
```

```python
#видим, что пропущенных значений не осталось
df_train['Age'].isna().sum()
```

```python
# перейдём к основной задаче
df_train
```

```python
df_surv = df_train.copy()
df_surv.dropna(subset='Survived', inplace=True)
df_pred = df_train[df_train['Survived'].isna()]
```

```python
df_surv
```

```python
df_pred
```

```python
Xtrain = df_surv.drop('Survived', axis=1)
ytrain = df_surv['Survived']
Xtest = df_pred.drop('Survived', axis=1)

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)
model.fit(Xtrain, ytrain)

y_pred = model.predict(Xtest)
print(y_pred[:10])
```

```python
pass_id = np.array(pass_id)
df_final = pd.DataFrame({'PassengerId': pass_id, 'Survived': y_pred})
df_final
```
